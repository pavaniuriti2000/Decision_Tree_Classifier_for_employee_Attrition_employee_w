# feature engineering

```python
#encoding of catogorical variables
#One hot encoding using pandas.getdummies
```

```python
df_fe = df_eda.copy()
```

```python
df_fe
```

```python
one_hot_var = ['BusinessTravel','Department','EducationField','JobRole','MaritalStatus']

for i in one_hot_var:
    x_t = pd.get_dummies(df[i],prefix=i,prefix_sep='_',drop_first=True)
    df_fe = df_fe.join(x_t)
    
    df_fe.drop(i,axis=1,inplace=True)
```

```python
df_fe
```

```python
df_fe.Attrition = df_fe.Attrition.apply(lambda x: 1 if x=='Yes' else 0)
```

```python
df_fe
```

```python
df_fe.Gender = df_fe.Gender.apply(lambda x: 1 if x=='Male' else 0)
```

```python
df_fe
```

```python
#now catogorical encoding is done
```

```python
x = df_fe.drop('Attrition',axis=1)
```

```python
x
```

```python
y= df_fe['Attrition']
```

```python
y
```

```python
from sklearn.model_selection  import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.25)
```
